## データオーグメンテーションを試してみよう

連合学習をするような仲間は集められないけどなんとか精度を底上げしたい...\
そんなとき，手軽に試せるのがData Augmentation（データ拡張）です．

今回は元の画像に様々な変換を加えていくことになるので，視覚的な確認を行うための方法を確認しておきましょう．

まずは例によってライブラリのインポートを行い，少数サンプルのMNISTを用意します．

```python
import torch
import torch.nn.functional as F
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms

mnist_train = datasets.MNIST("./data", train=True, download=False, transform=transforms.Compose([transforms.ToTensor()]))
mnist_test = datasets.MNIST("./data", train=False, transform=transforms.Compose([transforms.ToTensor()]))

n_samples = 100 
train_size = int(n_samples * 0.8) 
val_size = n_samples - train_size 

small_dataset, _ = torch.utils.data.random_split(mnist_train, [n_samples, len(mnist_train)-n_samples])

mnist_small_train, mnist_small_val = torch.utils.data.random_split(small_dataset, [train_size, val_size])

train_loader = DataLoader(mnist_small_train, batch_size=10, shuffle=True)
val_loader = DataLoader(mnist_small_val, batch_size=10, shuffle=False)
test_loader = DataLoader(mnist_test, batch_size=100, shuffle=False)
```

次に，train_loaderから読み込まれる80枚の画像を可視化する関数を作ります．

```python
import matplotlib.pyplot as plt

def show_data(train_loader):
    imgs = []
    W = 10
    H = 8
    fig = plt.figure(figsize=(H, W))
    for i, data in enumerate(train_loader):
        x, t = data
        for j in range(len(x)):
            ax = fig.add_subplot(H, W, (10*i)+(j+1), xticks=[], yticks=[])
            ax.imshow(x[j][0], cmap="gray")
```

```python
show_data(train_loader)
```

これで，trainloaderがロードする画像を可視化する関数show_dataができあがりました．\
この関数を使って変換結果を確認しながら，様々なオーグメンテーションを試してみましょう．

## データオーグメンテーションを行う方法
### よくある誤解
- 元の画像を回転させたり反転させたりしたものを沢山作り，画像フォルダに追加しておかなければならない

もちろんそのように行うこと自体は間違いではないのですが，実際はわざわざ水増ししたものを保存し直すことをする必要はありません．

ではどのようにデータを拡張すればよいのでしょうか？\
それにたいする１つの解答が**transform**です．

### transform
実はここまででMNISTのデータセットを呼び出す際に，transformという引数を与えていました．\
改めてスクリプトを確認してみましょう．

```python
mnist_train = datasets.MNIST("./data", train=True, download=False, transform=transforms.Compose([transforms.ToTensor()]))
mnist_test = datasets.MNIST("./data", train=False, transform=transforms.Compose([transforms.ToTensor()]))
```

transform=transforms.Compose([transforms.ToTensor()])\
という記述が確認できるかと思います．これは，「データセットに含まれる画像がDataLoaderにより読み込まれる際に，どのような変換を施すか」を表しています．\
transforms.Compose()の中にリスト形式で変換手法を並べておくことにより，それらの変換が順番に施されます．\
上記では，transforms.ToTensor()のみが記述されているため，「読み込んだ画像をTensor型にする」という変換のみが施されることになります．

transformで使える変換はToTensor()以外にも，様々なものが用意されています．下記で一例を示しておきます．
- RandomHorizontalFlip()
- RandomVerticalFlip()
- RandomRotation()
- RandomCrop()
- 他にも多数（興味ある方は[こちら](https://pytorch.org/vision/stable/transforms.html)】）

まずは，左右反転と上下反転を試してみましょう．\
下のセルで，一連のtransformを並べてComposeに格納しましょう．

```python
my_transform=transforms.Compose([transforms.ToTensor(),
                                 transforms.RandomHorizontalFlip(p=0.5),
                                 transforms.RandomVerticalFlip(p=0.5)])
```

ここで，p=0.5といったようなオプションが指定されていることにお気づきでしょうか？\
これは，その変換が施される確率を意味しています．

変換を確率的に行いながらミニバッチのサンプリングを繰り返すことで，モデルは変換前の画像も変換後の画像も経験することができます．\
わざわざ変換済みの画像をフォルダに保存し直す必要がないと述べたのは，このように確率的にデータオーグメンテーションを行うことが一般的だからです．

それでは，改めてMNISTのデータセットを呼び出してmy_transformを適用しましょう．\
データオーグメンテーションは訓練時の工夫ですので，trainデータのみに適用します．

```python
mnist_train = datasets.MNIST("./data", train=True, download=False, transform=my_transform)
```

そして，改めてデータを小数に分割し，train, valへの分割も併せて行います．

```python
n_samples = 100 
train_size = int(n_samples * 0.8) 
val_size = n_samples - train_size 

small_dataset, _ = torch.utils.data.random_split(mnist_train, [n_samples, len(mnist_train)-n_samples])

mnist_small_train, mnist_small_val = torch.utils.data.random_split(small_dataset, [train_size, val_size])

train_loader = DataLoader(mnist_small_train, batch_size=10, shuffle=True)
val_loader = DataLoader(mnist_small_val, batch_size=10, shuffle=False)
```

train_loaderにデータオーグメンテーションの機能が付加されているかどうか，先に定義したshow_data()で確認してみましょう．

```python
show_data(train_loader)
```

4や7など，上下左右にひっくり返っている文字が混ざっているのが確認できますね．\
ここで重要なのは，このような水増しが適切であるかどうかをきちんと確認することです．\
手書き数字認識の場合，ひっくり返っている数字を認識しなければならないケースはほとんどありません．また，6と9のように，ひっくり返すと意味が異なってしまうようなクラスが含まれていると，画像とラベルの対応に矛盾が生まれてしまいます．

今回の場合は，少しだけ回転を加える程度が落とし所ではないかと思われます．

自分で作るといっても，必ずしもゼロから機能を実装していく必要はありません．\
Salt & Pepper Noiseはscikit-imageというライブラリで用意されているので，それを使ったクラスをtransformで包み込むことにより，これまでと同様の使い方ができるようになります．

```python
from skimage.util import random_noise
import random

class salt_pepper_noise:
    def __init__(self, p=0.5):
        self.p = p
    
    def __call__(self, x):
        if random.random() > self.p:
            return torch.Tensor(random_noise(x, mode="salt", amount=0.05))
        else:
            return x
```

それでは，自作したsalt_pepper_noiseや，上で紹介できなかったオーグメンテーション手法も含めた様々なtransformを試してみましょう．

一番下のセルまで一気に実行すると，オーグメンテーションありの学習とテストが最後まで実行されます．\
色々なパターンで，精度がどのように変化するか確かめてみてください！

```python
# transforms.Composeの内容をいろいろ書き換えて遊んでみましょう
my_transform=transforms.Compose([transforms.ToTensor(),
                                 #transforms.RandomHorizontalFlip(p=0.5),
                                 #transforms.RandomVerticalFlip(p=0.5),
                                 #transforms.RandomRotation(10)
                                 #salt_pepper_noise(p=0.5)
                                ])

mnist_train = datasets.MNIST("./data", train=True, download=False, transform=my_transform)

n_samples = 100 
train_size = int(n_samples * 0.8) 
val_size = n_samples - train_size 

small_dataset, _ = torch.utils.data.random_split(mnist_train, [n_samples, len(mnist_train)-n_samples])

mnist_small_train, mnist_small_val = torch.utils.data.random_split(small_dataset, [train_size, val_size])

train_loader = DataLoader(mnist_small_train, batch_size=10, shuffle=True)
val_loader = DataLoader(mnist_small_val, batch_size=10, shuffle=False)

show_data(train_loader)
```

```python
# ネットワークの定義
class MLP(nn.Module):
    def __init__(self):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(28*28, 1000)
        self.fc2 = nn.Linear(1000, 10)
        
    def forward(self, x):
        x = x.view(-1, 28 * 28)
        x = self.fc1(x)
        x = F.relu(x)
        x = self.fc2(x)
        
        return F.softmax(x)

# 誤差関数の定義
criterion = nn.CrossEntropyLoss()

# モデルの定義
model = MLP()

# GPUを使う場合は，下記のコメントを外す
model = model.to("cuda")

# optimizer（勾配降下法のアルゴリズム）の準備
optimizer = optim.Adam(model.parameters())

# 学習ループ
epochs = 30 #ミニバッチのサンプリングが一巡 = 1 epoch

train_loss_list = [] # epoch毎のtrain_lossを保存しておくための入れ物
val_loss_list = [] # epoch毎のvalidation_lossを保存しておくための入れ物

loss_min = 100000 # validation_lossが小さくなった場合にのみモデルを保存しておくためのメモ

# ここからループ開始
for epoch in range(epochs):
    
    train_loss_add = 0 # 1エポック分の誤差を累積しておくための変数
    
    model.train() #学習モードであることを明示
   
    for i, data in enumerate(train_loader):
        
        x, t = data # ①データの読み込み
        
        #GPU環境で動かす際は，下記2行のコメントを外す
        x = x.to("cuda")
        t = t.to("cuda")
        
        predict = model(x) # ②順伝播計算
        
        loss = criterion(predict, t) # ③誤差の計算
        
        model.zero_grad()# 誤差逆伝播法のための準備
        loss.backward() # ④誤差逆伝播法による誤差の計算
        
        optimizer.step() # ⑤勾配を用いてパラメータを更新
        
        train_loss_add += loss.data # あとで平均を計算するために，誤差を累積しておく
        
    train_loss_mean = train_loss_add / int(len(mnist_small_train)/train_loader.batch_size) # 1epochでの誤差の平均を計算
    print("epoch" + str(epoch+1))
    print("train_loss:" + str(train_loss_mean))
    train_loss_list.append(train_loss_mean.cpu())# 1epoch毎の平均を格納しておく
    
    # validation
    model.eval() # 評価モード（学習を行わない）であることを明示
    
    val_loss_add = 0
    for i, data in enumerate(val_loader):
            
        x, t = data
        
        #GPU環境で動かす際は，下記2行のコメントを外す
        x = x.to("cuda")
        t = t.to("cuda")
        
        predict = model(x) # 順伝播計算
        
        loss = criterion(predict, t) # 誤差の計算
        val_loss_add += loss.data
        
    val_loss_mean = val_loss_add / int(len(mnist_small_val)/val_loader.batch_size)
    print("val_loss:" + str(val_loss_mean))
    val_loss_list.append(val_loss_mean.cpu())
    
    if val_loss_mean < loss_min: # 前に保存したモデルよりもvalidation lossが小さければ，モデルを保存する
        torch.save(model.state_dict(), "/takaya_workspace/self_study/deep_learning/federated_learning/models/best.model")
        print("saved best model!")
        loss_min = val_loss_mean # 今回保存したモデルのvalidation lossをメモしておく
```

```python
import matplotlib.pyplot as plt
plt.plot(train_loss_list, label="train") # train時のlossをplot
plt.plot(val_loss_list, label="val") # validation時のlossをplot
plt.xlabel("epoch") # X軸のラベルを設定
plt.ylabel("loss") # Y軸のラベルを設定
plt.legend() # 凡例を追加（plot時に指定したlabelが使われる）
```

```python
model = MLP() # テストに使うためのMLPを改めて定義

#GPU環境で動かす際は，下記のコメントを外す
model = model.to("cuda")

#高屋が予め用意した学習済みモデルを使う場合は，best.modelをtakaya.modelに書き換えてください
model_path = "/takaya_workspace/self_study/deep_learning/federated_learning/models/best.model"

model.load_state_dict(torch.load(model_path)) # 学習時に保存しておいたモデルのパラメータをコピー

model.eval() # 評価モードであることを明示

acc_count = 0 # 正解した数をカウントしておくための変数


for i, data in enumerate(test_loader):
    
    x, t = data
    
    #GPU環境で動かす際は，下記2行のコメントを外す
    x = x.to("cuda")
    t = t.to("cuda")
        
    predict = model(x) # 学習済みU-Netによる推論
    
    acc_count += (predict.argmax(axis=1) == t).sum().item()

print("Test Accuracy: " + str(acc_count / len(mnist_test)))
```
